```python
# Reading the saved links
links_df = pd.read_csv("Links.csv", index_col = 0)
links_df[links_df['Links'].isnull()].count
```

```python
# Dropping the NAN values from the data
df = links_df.dropna()
print(len(df))
# Converting into array and taking 500 links
videos = np.array(df['Links'][0:500])
print(len(videos))
# initialising the dictionary to store the extracted data from the links
videos_dictionary = {}
videos_dictionary['Video Link'] = {}
videos_dictionary['Video Title'] = {}
videos_dictionary['Views'] = {}
videos_dictionary['Uploaded Date'] = {}
videos_dictionary['Likes on Video'] = {}
videos_dictionary['Dislikes on Video'] = {}
```

Defining the function to scrap the required information from the links

```python
def scrap(url,i):
    
    print('Fetched date and time - ',datetime.now().strftime("%d/%m/%Y %H:%M:%S"))
    
    try :
        videos_dictionary['Video Link'].update({i : url})
    except:        
         videos_dictionary['Video Link'].update({i :''})
    
    driver.get(url)
    
    time.sleep(2)
    
    try :
        title =  driver.find_element_by_xpath('//*[@id="container"]/h1/yt-formatted-string')
        videos_dictionary['Video Title'].update({i : title.text})
    except :
        videos_dictionary['Video Title'].update({i : ''})
        
    try :
        views =  driver.find_element_by_xpath('//*[@id="count"]/yt-view-count-renderer')
        videos_dictionary['Views'].update({i : views.text})
    except :
        videos_dictionary['Views'].update({i : ''})
    
    try :
        date =  driver.find_element_by_xpath('//*[@id="date"]/yt-formatted-string')
        videos_dictionary['Uploaded Date'].update({i : date.text})
    except :
        videos_dictionary['Uploaded Date'].update({i : ''})
        
    try :
        likes =  driver.find_element_by_xpath('//*[@id="top-level-buttons"]/ytd-toggle-button-renderer[1]/a')
        videos_dictionary['Likes on Video'].update({i :likes.text})
    except :
        videos_dictionary['Likes on Video'].update({i : ''})
        
    try :
        dislikes =  driver.find_element_by_xpath('//*[@id="top-level-buttons"]/ytd-toggle-button-renderer[2]/a')
        videos_dictionary['Dislikes on Video'].update({i : dislikes.text})
    except :
        videos_dictionary['Dislikes on Video'].update({i : ''})
    
```

```python
# Scrapping the info 
for num in range (start, len(videos)):
    url = videos[num]
    print("=" * 40)
    print("Loop entered")
    try :
        print("getting link")
        driver.get(url)
    except :
        print("Not getting")
        continue
        

    time.sleep(2)
    print("=" * 40)  # Shows in terminal when a new airline is being scraped
    print("Scraping " + url)
    
    scrap(url,num)
    
print("*"*20,"Process Ended Successfully", "*"*20)  
```
